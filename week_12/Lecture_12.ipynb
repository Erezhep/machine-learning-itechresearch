# I.–¢–µ–æ—Ä–∏—è–ª—ã“õ —Ç–∞–ø—Å—ã—Ä–º–∞

## 1-—Ç–∞–ø—Å—ã—Ä–º–∞. –ú–µ—Ç—Ä–∏–∫–∞–ª–∞—Ä–¥—ã —Ç“Ø—Å—ñ–Ω–¥—ñ—Ä—É

–ë–µ—Ä—ñ–ª–≥–µ–Ω confusion matrix “Ø—à—ñ–Ω –∫–µ–ª–µ—Å—ñ –º–µ—Ç—Ä–∏–∫–∞–ª–∞—Ä–¥—ã –µ—Å–µ–ø—Ç–µ“£—ñ–∑:

|   | –ù–∞“õ—Ç—ã + | –ù–∞“õ—Ç—ã - |
|---|---------|---------|
|–ë–æ–ª–∂–∞–º + | 45 | 15 |
|–ë–æ–ª–∂–∞–º - | 5  | 435 |

–°“±—Ä–∞“õ—Ç–∞—Ä:
1. Accuracy
2. Precision
3. Recall
4. F1-score
5. “ö–∞–π –º–µ—Ç—Ä–∏–∫–∞ –±“±–ª –∂–∞“ì–¥–∞–π–¥–∞ –µ“£ –º–∞“£—ã–∑–¥—ã


````python
TP = 45
FN = 15
FP = 5
TN = 435
Accuracy = (TP + TN) / (TP + TN + FP + FN)
Precision = TP / (TP + FP)
Recall = TP / (TP + FN)
F1_score = 2 * (Precision * Recall) / (Precision + Recall)

print(f"Accuracy: {Accuracy}")
print(f"Precision: {Precision}")
print(f"Recall: {Recall}")
print(f"F1-score: {F1_score}")
````

–ï“£ –º–∞“£—ã–∑–¥—ã –º–µ—Ç—Ä–∏–∫–∞ –º–µ–Ω—ñ“£—à–µ F1-score

## 8-—Ç–∞–ø—Å—ã—Ä–º–∞. –ë–∞–∑–∞–ª—ã“õ –º–æ–¥–µ–ª—å

1. SMOTE class_weight-—Ç—ñ –∞–ª—ã–ø —Ç–∞—Å—Ç–∞“£—ã–∑
2. Logistic Regression “Ø–π—Ä–µ—Ç—ñ“£—ñ–∑
3. Accuracy, Recall, F1 —Å–∞–ª—ã—Å—Ç—ã—Ä—ã“£—ã–∑
4. “ö–æ—Ä—ã—Ç—ã–Ω–¥—ã –∂–∞—Å–∞“£—ã–∑

````python
# Install missing libraries (Colab-safe)
!pip install -q imbalanced-learn

import numpy as np
import pandas as pd

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, classification_report
)

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

# ----------------
# 1. Generate imbalanced dataset
# ----------------
X, y = make_classification(
    n_samples=5000,
    n_features=20,
    n_informative=6,
    n_redundant=2,
    weights=[0.95, 0.05],  # imbalance
    random_state=42
)

print("Class distribution:", np.bincount(y))

# ----------------
# 2. Train-test split
# ----------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
````

````python
# Logistic Regression (base model)
model = LogisticRegression()

# train model
model.fit(X_train, y_train)

# predict
y_pred = model.predict(X_test)

# metrics
acc = accuracy_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {acc:.3f}")
print(f"Recall: {rec:.3f}")
print(f"F1-score: {f1:.3f}")
````

## 9-—Ç–∞–ø—Å—ã—Ä–º–∞. SMOTE ”ô—Å–µ—Ä—ñ

3. Recall –∂”ô–Ω–µ ROC-AUC —Å–∞–ª—ã—Å—Ç—ã—Ä—ã“£—ã–∑
4. “ö–∞–π—Å—ã—Å—ã minority class “Ø—à—ñ–Ω –∂–∞“õ—Å—ã?

````python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

model_smote = LogisticRegression()
model_smote.fit(X_train_smote, y_train_smote)

y_pred_smote = model_smote.predict(X_test)
````

## 10-—Ç–∞–ø—Å—ã—Ä–º–∞. Class weight ”ô—Å–µ—Ä—ñ

````python
weights = [None, 'balanced']

for weight in weights:
  mod = LogisticRegression(
      class_weight=weight,
      random_state=42
  )
  mod.fit(X_train, y_train)
  y_pred = mod.predict(X_test)

  print(f"class_weight: {weight} ----> f1_score: {f1_score(y_test, y_pred)}")
````

````python
# base model
model = LogisticRegression(max_iter=1000, solver="liblinear")

# params
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'penalty': ['l1', 'l2']
}

# GridSearch - F1
grid_f1 = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='f1',
    cv=5
)
grid_f1.fit(X_train, y_train)
print("Best F1 params:", grid_f1.best_params_)
print("Best F1 score:", grid_f1.best_score_)

print(f"\n{'-'*60}\n")

# 4) GridSearchCV - ROC-AUC
grid_roc = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    scoring='roc_auc',
    cv=5
)

grid_roc.fit(X_train, y_train)
print("Best ROC-AUC params:", grid_roc.best_params_)
print("Best ROC-AUC score:", grid_roc.best_score_)
````

F1-score –±–æ–π—ã–Ω—à–∞ –µ“£ –∂–∞“õ—Å—ã –ø–∞—Ä–∞–º–µ—Ç—Ä–ª–µ—Ä `C=0.01, penalty='l2'`, score —Ç”©–º–µ–Ω (0.25) —Å–µ–±–µ–±—ñ –æ–ª —Ç–µ–∫ minority class-“õ–∞ –Ω–∞–∑–∞—Ä –∞—É–¥–∞—Ä–∞–¥—ã –∂”ô–Ω–µ –∞–∑ –∫–ª–∞—Å—Å—Ç—ã“£ “õ–∞—Ç–µ—Å—ñ–Ω –∞–∑–∞–π—Ç—É“ì–∞ –±–∞“ì—ã—Ç—Ç–∞–ª“ì–∞–Ω, –∞–ª ROC-AUC –±–æ–π—ã–Ω—à–∞ `C=0.1, penalty='l1'`, score –∂–æ“ì–∞—Ä—ã (0.82) —Å–µ–±–µ–±—ñ –æ–ª –±–∞—Ä–ª—ã“õ –∫–ª–∞—Å—Ç–∞—Ä–¥—ã –∂–∞–ª–ø—ã –∞–∂—ã—Ä–∞—Ç—É “õ–∞–±—ñ–ª–µ—Ç—ñ–Ω ”©–ª—à–µ–π–¥—ñ; –∞–π—ã—Ä–º–∞—à—ã–ª—ã“õ –æ—Å—ã: F1 –Ω–∞“õ—Ç—ã minority class “Ø—à—ñ–Ω, ROC-AUC –∂–∞–ª–ø—ã performance “Ø—à—ñ–Ω –∂–∞“õ—Å—ã—Ä–∞“õ.
