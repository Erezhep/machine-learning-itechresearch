Екі 3×3 матрица құрып, олардың қосындысын, айырмасын және көбейтіндісін табыңыз.


```python
import numpy as np

A = np.array([
    [5, 12, 3],
    [4, 2, 6],
    [7, 8, 34]
])

B = np.array([
    [3, 0, 7],
    [8, 5, 4],
    [3, 2, 2]
])

print(f"A + B = \n{A + B}")
print(f"A - B = \n{A - B}")
print(f"A @ B = \n{A @ B}")
```

3 өлшемді екі вектордың скаляр көбейтіндісін және олардың арасындағы бұрышты есептеңіз.image.png

```python
a = np.array([3, 4, 1])
b = np.array([3, 9, 4])
scal = np.dot(a, b)
print("Скаляр көбейтінді:", scal)

m1 = np.linalg.norm(a)
m2 = np.linalg.norm(b)
angle = np.arccos(scal / (m1 * m2))
print("Бұрыш:", angle)
```

Берілген матрицаның детерминантын, кері матрицасын және өз мәндерін табыңыз.

---
image.png



```python
a = np.array([
    [1, 2],
    [3, 4]
])

print("Анықтауыш:", np.linalg.det(a))
print("Кері матрица:", np.linalg.inv(a))
print("Өз мәндері:", np.linalg.eig(a))
```

np.random.randn() арқылы 1000 элементтік үлгі алыңыз және орташа, дисперсия мен стандартты ауытқуды есептеңіз.

```python
r = np.random.randn(1000)
print("Орта мәні:", r.mean())
print("Дисперсия:", r.var())
print("Стандартты ауытқу:", r.std())
```

Берілген тізім: [5, 10, 15, 20, 25] — үшін Min-Max және Z-score нормализациясын орындаңыз.

```python
l = np.array([5, 10, 15, 20, 25])
mini = np.min(l)                             # x - x_max
                                            # x_max - x_min
maxi = np.max(l)
min_max = (l - mini) / (maxi - mini)
print("Mini-Max:", min_max)
z_score = (l - np.mean(l)) / np.std(l)
print("z-score:", z_score)
```
