**3-Дәріс тапсырмасы:**

Студенттердің аты мен бағалары бар сөздік жасап, ең жоғары бағаны тап.


```python
# Студенттің аты мен бағалары
students = {
    "Aruzhan": [85, 92, 78, 88, 90],
    "Bek": [95, 87, 91, 90, 93, 89],
    "Dias": [76, 80, 79, 85],
    "Dinara": [88, 90, 84, 79, 92],
    "Madi": [70, 75, 80, 85, 90, 95, 10],
    "Sherkhan": [66, 74, 82, 90, 88],
    "Sultan": [93, 97, 89, 85, 91]
}
```

```python
# Әр студенттің ең үздік бағаларын іздейміз
best_scores = {name: max(scores) for name, scores in students.items()}
best_scores
```

```python
# Ең жоғары балл және оның иесін табу
best = max(best_scores.items(), key=lambda x: x[1])
print(f"Ең жоғары бал иесі: {best[0]} ({best[1]})")
```
