**Тапсырма 1:** CSV Файлын Оқу және Тізімге Сақтау
1.   Жүктелген Salaries.csv файлын with open(...) операторын қолданып оқыңыз.
2.   Файлдың тек алғашқы 10 деректер жолын (data_rows атты тізімге) сақтаңыз. Тақырып жолын (header) санамаңыз.
3.   data_rows тізімінің ұзындығын (len()) және бірінші элементін (бірінші қызметкердің дерегі) экранға шығарыңыз.


```python
data_rows = []
with open("Salaries.csv", "r", encoding="utf-8") as f:
  for _ in range(11):
    if _ == 0:
      f.readline()
      continue
    data_rows.append(f.readline())
print(f"len(data_rows) -> {len(data_rows)}")
print(data_rows[0])
```

**Тапсырма 2:** Кортеж және Сөздік – Алғашқы 3 Адамның Атын Сақтау
1.   data_rows тізімінен алғашқы 3 қызметкердің Атын (EmployeeName) алып, оларды Кортежге (first_three_names) сақтаңыз.
2.   Осы 3 қызметкердің әрқайсысы үшін аты-жөнін (EmployeeName) Кілт ретінде, ал Жұмыс Атағын (JobTitle) Мән ретінде пайдаланып, Сөздік (employee_titles) құрыңыз.
3.   Кортежді және Сөздікті экранға шығарыңыз.


```python
# Id,EmployeeName,JobTitle,BasePay,OvertimePay,OtherPay,Benefits,TotalPay,TotalPayBenefits,Year,Notes,Agency,Status

# Алғашқы 3 адамның есімі
first_three_names = tuple([data_rows[i].split(",")[1] for i in range(3)])

# Алғашқы 3 адамның жұмысы
first_three_jobs = tuple([data_rows[i].split(",")[2]  for i in range(3)])

# Жаңа сөздің есімі: жұмысы
employee_titles = dict(zip(first_three_names, first_three_jobs))

# Экранға шығару
print(f"first_three_names -> {first_three_names}")
print(f"employee_titles -> {employee_titles}")
```

**Тапсырма 3:** Тізімдік Өрнектер (List Comprehension)
1.    data_rows тізіміндегі 10 қызметкердің барлығының Базалық Жалақысын (BasePay) мәтіндік (str) түрде алып, оларды тізімдік өрнек (List Comprehension) арқылы base_pays_str атты жаңа тізімге жинаңыз. Ескерту: Бұл жерде түрлендіру (float-қа) немесе сүзгілеу (filtering) қажет емес, тек мәтіндік деректі алу жеткілікті.
2.    Жасалған base_pays_str тізімін экранға шығарыңыз.


```python
# Id,EmployeeName,JobTitle,BasePay,OvertimePay,OtherPay,Benefits,TotalPay,TotalPayBenefits,Year,Notes,Agency,Status

 # row.split(",")[3] - алынған қатарды белгі бойынша бөліп, сол тізімнен 4-інші тұрған элементті алады
 # if row.split(",")[3].replace(".", "").isdigit() - тізімде жұмыс атауында , белгісі бар ол өз кезегінде жұмыс ақысының орнына жұмыс орнын екіге бөліп шығарып береді
 # else row.split(",")[4] - егер шынымен солай болса онда 4 емес 5-інші тұрған элементті алады

base_pays_str = [row.split(",")[3] if row.split(",")[3].replace(".", "").isdigit() else row.split(",")[4] for row in data_rows]
base_pays_str
```

**Тапсырма 4:** Жиын (Set) – Бірегей Атақтарды Табу
1.   data_rows тізіміндегі 10 қызметкердің барлығының Жұмыс Атақтарын (JobTitle) тізімдік өрнексіз, қарапайым for циклін пайдаланып алыңыз.
2.   Бұл тізімді Жиын (unique_titles) құрылымына түрлендіріңіз.
3.   Жиынның барлық элементтерін экранға шығарыңыз. Нелікwтен жиын пайдалы екенін қысқаша түсіндіріңіз (2 сөйлем).


```python
jobs_title = []
for row in data_rows:
  jobs_title.append(row.split(",")[2])
unique_titles = set(jobs_title)
print(*unique_titles, sep="\n")

"""
  1.Тізімді жиынға түрлендіру арқылы біз қайталаныт элементтерден құтылып, уникальды қайталанбайтын
  элементтері бар жаңа жиын аламыз.

  2.Толық тізімге анализ жасағанда біз қандай да бір қатарды топтай аламыз белгілі бір элементтер арқылы

"""
```
