**About Dataset**

One way to understand how a city government works is by looking at who it employs and how its employees are compensated. This data contains the names, job title, and compensation for San Francisco city employees on an annual basis from 2011 to 2014.


Зерттеу нәтижелері төменде жазылған

```python
import kagglehub
import pandas as pd
from kagglehub import KaggleDatasetAdapter
```

```python
# Load dataset from kaggle
file_path = "Salaries.csv"  # SF - San Francisco

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "kaggle/sf-salaries",
    file_path,
)
```

```python
emp_sal = df[['EmployeeName', 'BasePay', 'OvertimePay', 'OtherPay', 'TotalPay', 'Year']]
emp_sal.head()
```

```python
emp_sal.info()
```

```python
# String типті санға (float) түрлендіру
emp_sal['BasePay'] = pd.to_numeric(emp_sal['BasePay'], errors='coerce')
emp_sal
```

```python
emp_sal['BasePay'].fillna(emp_sal['BasePay'].mean(), inplace=True)
```

```python
emp_sal['BasePay'].isna().sum()
```

```python
emp_sal['OvertimePay'] = pd.to_numeric(emp_sal['OvertimePay'], errors='coerce')
emp_sal['OtherPay'] = pd.to_numeric(emp_sal['OtherPay'], errors='coerce')
emp_sal[['OvertimePay', 'OtherPay']].isna().sum()
```

```python
emp_sal['OvertimePay'].fillna(emp_sal['OvertimePay'].mean(), inplace=True)
emp_sal['OtherPay'].fillna(emp_sal['OtherPay'].mean(), inplace=True)
emp_sal.isna().sum()
```

Осы жерге дейін жұмысым өзіме керек бағандарды таңдап, оларды тазалау болды. Бос орындарды анықтап оларды орташа мәнмен толтырдым

# Төмендегі зерттеу бойынша қорытынды
1.  Базалық тұрақты орташа төлем 2011-2014 жылдары 63595 доллардан 66564 долларға дейін көтерілген
2.  Қосымша жұмыс уақыты үшін баға көтерілді деп айтуға негіз жоқ, себебі бұл адамдардың бір ситуацияға, мерекелік күндерде болмаса өз бетінше көп табамын деген оймен көп уақыт жұмыс істеуіне байланысты яғни жай ғана қосымша табыс
3.  Өзге табыс түрлері бойынша керісінше 3617 доллардан 3505 долларға төмендеген, яғни бұл компания төлейтін оклад деп есептесек ондай компания табысы 2011-2014 жылдары төмендегені байқалады себебі оклад бұл түскен пайданың бір бөлігін жұмысшыларға бөліп беру

```python
emp_sal.groupby(by='Year')[['BasePay', 'OvertimePay', 'OtherPay', 'TotalPay']].mean()
```

```python
group_by_name = emp_sal.groupby(by='Year')['EmployeeName'].count()
group_by_name
```

```python
group_by_salary = emp_sal.groupby(by = 'Year')['TotalPay'].sum()
group_by_salary
```

```python
salemp = pd.merge(group_by_name, group_by_salary, on = 'Year')
salemp['EverySalary'] = round(salemp['TotalPay'] / salemp['EmployeeName'], 2)
salemp['PercentEverySalary'] = round(salemp['EverySalary'] / salemp.iloc[0, 2], 2)
salemp
```
