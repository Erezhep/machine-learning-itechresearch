## Тапсырма 2: Pandas Series және DataFrame Операциялары

Келесі операцияларды орындау арқылы Pandas Series және DataFrame-лермен жұмыс жасауды үйреніңіз:

1. **'Age'** бағанынан ғана тұратын **Series** құрып, оның типі мен орташа мәнін (екі ондық таңбаға дейін) шығарыңыз.


```python
titanic_age = df['Age']
print(f"Type: {titanic_age.dtype}\nMean: {titanic_age.mean()}")
```

2. Тек **'Name', 'Age', 'Fare'** бағандарынан тұратын **жаңа DataFrame** құрып, оның алғашқы 5 жолын көрсетіңіз.


```python
# naf - Name, Age, Fare
naf = df[['Name', 'Age', 'Fare']]
naf
```

3. **'Age'** бағанын қолданып, келесі интервалдар бойынша **'Age_Group'** (Жас тобы) атты жаңа баған қосыңыз: $(0, 18], (18, 35], (35, 60], (60, 100]$. 'Name', 'Age' және 'Age_Group' бағандарының алғашқы 10 жолын көрсетіңіз.


```python
bins = [0, 18, 35, 60, 100]
labels = ["0-18", "19-35", "36-60", "61-100"]

naf['Age_Group'] = pd.cut(naf["Age"], bins=bins, labels=labels, right=True)
naf[['Name', 'Age', 'Age_Group']].head(10)
```

4. **'Fare'** бағанының мәндерін екі есе арттырып, **'Fare_Doubled'** атты жаңа баған қосыңыз. 'Name', 'Fare', 'Fare_Doubled' бағандарының алғашқы 5 жолын көрсетіңіз.


```python
naf['Fare_Doubled'] = naf['Fare'] * 2
naf[['Name', 'Fare', 'Fare_Doubled']].head(5)
```

## Тапсырма 4: Логикалық Шарттармен Сүзу

DataFrame-ді әртүрлі шарттар бойынша сүзуді орындаңыз:

1. **Тіршілік қалғандарды** (`Survived == 1`) сүзіңіз. Олардың жалпы санын және **'Name', 'Age', 'Sex', 'Pclass'** бағандары бойынша алғашқы 5 жолын көрсетіңіз.


```python
survived = df[df['Survived'] == 1][['Name', 'Age', 'Sex', 'Pclass']]
print(f"Жалпы саны: {len(survived)}")
survived.head()
```

2. **Екі шартты (& - AND)** қолданып, **жынысы әйел** және **1-класта** болған жолаушыларды сүзіңіз. Санын және **'Name', 'Age', 'Fare'** бағандары бойынша алғашқы 5 жолын көрсетіңіз.


```python
female_first_class = df[(df['Sex'] == 'әйел') & (df['Pclass'] == 1)]
print(len(female_first_class))
female_first_class[['Name', 'Age', 'Fare']].head()
```

3. **НЕМЕСЕ шартын (| - OR)** қолданып, **10 жасқа дейінгі** балаларды немесе **60 жастан асқан** кәрілерді сүзіңіз. Санын және **'Name', 'Age'** бағандары бойынша алғашқы 10 жолын көрсетіңіз.


```python
childs_or_olds = df[(df['Age'] <= 10) | (df['Age'] > 60)]
print(len(childs_or_olds))
childs_or_olds[['Name', 'Age']].head(10)
```

4. **`isin()`** функциясын қолданып, **'Embarked'** (порт) бағанындағы мәні **'S'** немесе **'C'** болып табылатын жолаушыларды сүзіңіз. Санын және **'Name', 'Embarked'** бағандары бойынша алғашқы 5 жолын көрсетіңіз.


```python
pas_s_c = df[df['Embarked'].isin(['S', 'C'])]
print(len(pas_s_c))
pas_s_c[['Name', 'Embarked']].head()
```

5. **Үш шартты** қолданып, билетінің құны **100-ден жоғары**, **1-класқа** жататын және **аман қалған** жолаушыларды сүзіңіз. Санын және **'Name', 'Fare', 'Pclass'** бағандары бойынша алғашқы 5 жолын көрсетіңіз.

```python
df[(df['Fare'] > 100) & (df['Pclass'] == 1) & (df['Survived'] == 1)][['Name', 'Fare', 'Pclass']].head()
```

## Тапсырма 5: Топтастыру (`Groupby`)

Деректерді топтастыру және агрегаттау арқылы мәліметтерді талдаңыз:

1. **'Sex'** бағаны бойынша топтастырып, **'Survived'** бағанының **орташа мәнін** (тіршілік қалу үлесін) табыңыз. Нәтижені және әйелдер мен ерлердің пайыздық үлесін жеке көрсетіңіз.


```python
df.groupby(by='Sex')['Survived'].mean()
```

2. **'Pclass'** және **'Sex'** бағандары бойынша топтастырып, **'Survived'** бағанының **орташа мәнін** шығарыңыз.


```python
df.groupby(by=['Sex', 'Pclass'])['Survived'].mean()
```

3. **'Pclass'** бойынша топтастырып, **'Age'** бағаны үшін **орташа мәнді (`mean`), минималды (`min`), максималды (`max`)** және **санақты (`count`)** есептеңіз.


```python
df.groupby(by='Pclass')['Age'].agg(['mean', 'min', 'max', 'count'])
```

4. **'Embarked'** (порт) бойынша топтастырып, **'Age'** және **'Fare'** бағандарының **орташа мәндерін** табыңыз.


```python
df.groupby(by='Embarked')[['Age', 'Fare']].mean()
```

5. **'Pclass'** бойынша топтастырып, әр кластағы **жолаушылардың санын (`size`)** есептеңіз. Нәтижені және барлық жолаушылардың жалпы санын көрсетіңіз.

```python
df.groupby(by='Pclass')['Pclass'].count()
```

## Тапсырма 7: Кестелерді Біріктіру (`Merge` және `Concat`)

Қосымша DataFrame-дер құрып, оларды әртүрлі әдістермен біріктіріңіз:

*(Бастапқы кодтағыдай, біріктіру үшін `passengers`, `tickets`, `survival_info` және `df_family` сияқты шағын DataFrame-дерді құрыңыз)*

passengers = pd.DataFrame({
    'PassengerId': [1, 2, 3],
    'Name': ['Braund', 'Cumings', 'Heikkinen']
})
tickets = pd.DataFrame({
    'PassengerId': [1, 2, 4],
    'Ticket': ['A123', 'B456', 'C789']
})
survival_info = pd.DataFrame({
    'PassengerId': [1, 2, 3, 5],
    'Survived': [0, 1, 1, 0]
})


1. **`merge()`** функциясын қолданып, **`passengers`** және **`tickets`** DataFrame-лерін **`PassengerId`** бағаны бойынша **INNER JOIN** (ішкі біріктіру) арқылы біріктіріңіз.


```python
passengers = pd.DataFrame({ 'PassengerId': [1, 2, 3], 'Name': ['Braund', 'Cumings', 'Heikkinen'] })
tickets = pd.DataFrame({ 'PassengerId': [1, 2, 4], 'Ticket': ['A123', 'B456', 'C789'] })
survival_info = pd.DataFrame({ 'PassengerId': [1, 2, 3, 5], 'Survived': [0, 1, 1, 0] })
```

```python
pt = pd.merge(passengers, tickets, how='inner', on='PassengerId')
pt
```

2. **`merge()`** функциясын қолданып, **`passengers`** және **`survival_info`** DataFrame-лерін **`PassengerId`** бағаны бойынша **LEFT JOIN** (сол жақ біріктіру) арқылы біріктіріңіз.


```python
ps = pd.merge(passengers, survival_info, how='left', on='PassengerId')
ps
```

3. **`df`** негізгі DataFrame-нің **'Name'** және **'Age'** бағандарынан тұратын алғашқы 3 жолды (`df1`) және соңғы 3 жолды (`df2`) құрып, оларды **`concat()`** арқылы **тігінен (жолдарды қосу)** біріктіріңіз.


```python
# ht => Head, Tail
ht = pd.concat([df.head(3), df.tail(3)])
ht
```

4. **`df`** негізгі DataFrame-нің **'Name', 'Age'** (алғашқы 4 жол) және **'Sex', 'Fare'** (алғашқы 4 жол) бағандарынан тұратын екі бөлікті **`concat()`** арқылы **көлденеңінен (бағандарды қосу)** біріктіріңіз.


```python
pd.concat([df[['Name', 'Age']].head(4), df[['Sex', 'Fare']].head(4)], axis=1)
```

5. **'Pclass'** және **'Sex'** бағандары бойынша біріктіруге арналған **жаңа `df_family`** DataFrame-ін құрыңыз. **`df`**-тің алғашқы 10 жолын **`df_family`**-мен **`Pclass`** және **`Sex`** бағандары бойынша **LEFT JOIN** арқылы біріктіріңіз.

```python
df_family = df[['Pclass', 'Sex']]
pd.merge(df.head(10), df_family, how='left', on=['Pclass', 'Sex'])
```

5. **`fillna()`**: **'Age'** бағанындағы **NaN** мәндерін **сол бағанның медианасымен** толтырыңыз. Толтырғаннан кейінгі **NaN мәндердің санын** шығарыңыз.
